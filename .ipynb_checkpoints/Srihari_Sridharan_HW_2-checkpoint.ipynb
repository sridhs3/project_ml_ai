````python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.dpi':(100)})
sns.set_style("darkgrid")
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
````

Fetching data from OpenML database (now comes packaged with scikit-learn)

````python
student_performance = fetch_openml(data_id=42351, as_frame=True)
df = student_performance.data
print(f"Data has {df.shape[0]} rows and {df.shape[1]} columns.")
print(f"Number of rows with missing data : {df.isnull().sum(axis=1).sum()}")
````

To simplify, I will be ommiting the G1, and **using only G2 as label**. 

````python
#snippet to generate labels depending on binay or multi-class classification
def make_labels(y, evaluation='multi_class'):
    if evaluation == 'binary':
        binary_y = (y >=12).astype(int)
        return binary_y
    
    elif evaluation == 'multi_class':
        mc_y = y.copy()
        mc_y[y <4] = 5
        mc_y[y.between(3,9)] = 4
        mc_y[y.between(9,13)] = 3
        mc_y[y.between(13,17)] = 2
        mc_y[y.between(17,20)] = 1
        return mc_y

# Snippet to vectorise the features using LabelEncoder, or One Hot Encoder. Just extended to make sure if we split the data into test and train, store the label encoding for consistent encoding. 
def categorical_encode(df, label_encoders=None):
    encoded_df = df.copy()
    categorical_vars = df.select_dtypes(include=['object']).columns # we are choosing only the non-numeric columns for categorical encoding
    if label_encoders is None: #if train or unsplit data
        label_encoders = {}
        for var in categorical_vars:
            le = OneHotEncoder()
            encoded_df[var] = le.fit_transform(df[var])
            label_encoders[var] = le
        return encoded_df, label_encoders
    else:
        for var in categorical_vars:
            le = label_encoders[var]
            encoded_df[var] = le.transform(df[var])
        return encoded_df

def categorical_onehot_encode(df, label_encoders=None):
    encoded_df = df.copy()
    categorical_vars = df.select_dtypes(include=['object']).columns # we are choosing only the non-numeric columns for categorical encoding
    if label_encoders is None: #if train or unsplit data
        label_encoders = {}
        for var in categorical_vars:
            le = OneHotEncoder()
            transformed = le.fit_transform(df[[var]]).toarray()
            cols = [f"{var}_{category}" for category in le.categories_[0]] #create new columns ofr the one hot vecs
            encoded_df = encoded_df.drop(var, axis=1) #remove the original label now that we have one hot
            encoded_df[cols] = transformed
            label_encoders[var] = le
        return encoded_df, label_encoders
    else:
        for var in categorical_vars:
            le = label_encoders[var]
            transformed = le.transform(df[[var]]).toarray()
            cols = [f"{var}_{category}" for category in le.categories_[0]]
            encoded_df = encoded_df.drop(var, axis=1)
            encoded_df[cols] = transformed
        return encoded_df
````

### Binary Classification

We will consider this case as a straightforward pass/fail evaluation. We assume that **minimum percentage to pass the class is 60%** which corresponds to getting atleast 12 in the overall grade.

````python
#Labels
y = df['G2']
binary_y = make_labels(y, evaluation='binary') #get labels for binary classification
#multi_y = make_labels(y, evaluation='multi_class') #get labels for multi_class classification

#num of entries in each bin
binary_unique, binary_counts = np.unique(binary_y, return_counts=True)

#Features
# we are doing one hot for categorical features
X = df.drop(['G1', 'G2'], axis=1)
encoded_feats, encoded_feats_encoders = categorical_onehot_encode(X)
````

````python
# splitting the data into train (80%) and test (20%) sets. We will use the existing train_test_split from the scikit-learn
X_train, X_test, y_train, y_test = train_test_split(encoded_feats, binary_y, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
````

# Task 1 : Decision Tree Classifier

We will use skleearn's implementation

````python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# we'll evaluate the model
y_pred = clf.predict(X_test)
report = classification_report(y_test, y_pred, output_dict=True)
accuracy = report['accuracy']
f1_score = report['weighted avg']['f1-score'] 
print(f"Accuracy: {accuracy:.2f}\nF1 Score: {f1_score:.2f}")
````

Test Performance

Let's look at some feature importance as well, which is an important benefit of tree classifiers.

````python
def get_feature_importance(train_x, train_y):    
    rf = DecisionTreeClassifier()
    rf.fit(train_x, train_y)
    y_pred = rf.predict(train_x)
    importance = rf.feature_importances_
    feature_importances = pd.DataFrame(rf.feature_importances_, 
                                       index = train_x.columns, 
                                       columns=['importance']).sort_values('importance',ascending=False)
    return feature_importances
````

````python
feature_importance = get_feature_importance(X_train, y_train)
_ = feature_importance[:10].plot.barh(y='importance', figsize = (8,6))
````

First is the obvious interpretation the presence of failures features. This is basically number of past class failures. This suggests that people who have failed in the past, haven't taken much effort to improve.

We'll tweak the parameters for split criterion, maximum depth of the tree (0-10), minimum number of samples required to split and minimum number of samples required to be at a leaf node. We will `GridSearchCV` from sklearn to evaluate for each hyper-parameter. 

````python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 2, 3, 4, 5, 6, 7, 8],
    'min_samples_split': [2, 3, 4, 5, 6],
    'min_samples_leaf': [1, 2, 4]
}

clf = DecisionTreeClassifier(random_state=31)
grid = GridSearchCV(estimator=clf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=0, scoring='accuracy')
grid.fit(X_train, y_train)


print("Best parameters : ", grid.best_params_)
print("Best accuracy : ", grid.best_score_)

# we'll plot 
results = pd.DataFrame(grid.cv_results_)
results['criterion'] = results.params.apply(lambda x: x['criterion'])
results['min_samples_split'] = results.params.apply(lambda x: x['min_samples_split'])
results['min_samples_leaf'] = results.params.apply(lambda x: x['min_samples_leaf'])
if None in param_grid['max_depth']:
    results['max_depth'] = results.params.apply(lambda x: x['max_depth'] if x['max_depth'] is not None else 'None')

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

sns.boxplot(ax=axes[0, 0], data=results, x='max_depth', y='mean_test_score')
axes[0, 0].set_title('Accuracy vs Max Depth')
axes[0, 0].set_xlabel('Max Depth')
axes[0, 0].set_ylabel('Mean Test Score')

sns.boxplot(ax=axes[0, 1], data=results, x='criterion', y='mean_test_score')
axes[0, 1].set_title('Accuracy vs Criterion')
axes[0, 1].set_xlabel('Criterion')
axes[0, 1].set_ylabel('Mean Test Score')

sns.boxplot(ax=axes[1, 0], data=results, x='min_samples_split', y='mean_test_score')
axes[1, 0].set_title('Accuracy vs Min Samples Split')
axes[1, 0].set_xlabel('Min Samples Split')
axes[1, 0].set_ylabel('Mean Test Score')

sns.boxplot(ax=axes[1, 1], data=results, x='min_samples_leaf', y='mean_test_score')
axes[1, 1].set_title('Accuracy vs Min Samples Leaf')
axes[1, 1].set_xlabel('Min Samples Leaf')
axes[1, 1].set_ylabel('Mean Test Score')

plt.tight_layout()
plt.show()
````

- We can see that maximum depth of the tree provides a better output with depth.
- Gini criterion has a better performance.
- Minimum number of samples required to split a node affects the accuracy.
- Minimum number of samples to split a node has improved with value 4.

**Overall by tweaking values, the accuracy has improved to 0.685 from 0.64**

# Task 3 : Evaluation Metrics

To compare the models we will evaluate using the following metrics : accuracy (train and test), F1 score, and AUC

- Accuracy is a general estimate of how many correct among the lot. If the classes are balanced (similar false positives and false negatives), this metric suffices. Otherwise, if there's a single majority class say 80%, the classifier can ignore the 20% leading to poor recall.
- For **imbalanced data**, a harsher evaluation is F1 score. False negatives and false positives have different consequences. This balances for better precision and recall. **In balanced dataset, this wont provide anything more than accuracy.** (Shouldn't give much in our case, but we'll anyway implement it. 
- AUC is another good metric suitable for imbalanced data, and quanitifes **ability to distinguish classes** (Suited for binary classification). This is area under the curve when plotting ROC curve. A random guessing classifier will have diagonal as curve, wheras a perfect classifier will have 1.0 AUC.

````python
# all threee classifiers
rf = RandomForestClassifier(random_state=42)
gb = GradientBoostingClassifier(random_state=42)
dt = DecisionTreeClassifier(criterion='gini', max_depth=6, min_samples_leaf=4, min_samples_split=2, random_state=42) #optimised decision tree

scoring_metrics = ['accuracy', 'f1', 'roc_auc']

dt_scores = cross_validate(dt, encoded_feats, binary_y, cv=k, scoring=scoring_metrics, return_train_score=True)
rf_scores = cross_validate(rf, encoded_feats, binary_y, cv=k, scoring=scoring_metrics, return_train_score=True)
gb_scores = cross_validate(gb, encoded_feats, binary_y, cv=k, scoring=scoring_metrics, return_train_score=True)

#accuracies
dt_avg_test_accuracy = np.mean(dt_scores['test_accuracy'])
rf_avg_test_accuracy = np.mean(rf_scores['test_accuracy'])
gb_avg_test_accuracy = np.mean(gb_scores['test_accuracy'])
dt_avg_train_accuracy = np.mean(dt_scores['train_accuracy'])
rf_avg_train_accuracy = np.mean(rf_scores['train_accuracy'])
gb_avg_train_accuracy = np.mean(gb_scores['train_accuracy'])

dt_avg_f1 = np.mean(dt_scores['test_f1'])
rf_avg_f1 = np.mean(rf_scores['test_f1'])
gb_avg_f1 = np.mean(gb_scores['test_f1'])

dt_avg_auc = np.mean(dt_scores['test_roc_auc'])
rf_avg_auc = np.mean(rf_scores['test_roc_auc'])
gb_avg_auc = np.mean(gb_scores['test_roc_auc'])
````

````python
print(f"Decision Tree average test accuracy\t\t: {dt_avg_test_accuracy:.2f}")
print(f"Random Forest average test accuracy\t\t: {rf_avg_test_accuracy:.2f}")
print(f"Gradient Boosting average test accuracy\t\t: {gb_avg_test_accuracy:.2f}\n")

print(f"Decision Tree average train accuracy\t\t: {dt_avg_train_accuracy:.2f}")
print(f"Random Forest average train accuracy\t\t: {rf_avg_train_accuracy:.2f}")
print(f"Gradient Boosting average train accuracy\t: {gb_avg_train_accuracy:.2f}\n\n")

print(f"Decision Tree average F1 score\t\t\t: {dt_avg_f1:.2f}")
print(f"Random Forest average F1 score\t\t\t: {rf_avg_f1:.2f}")
print(f"Gradient Boosting average F1 score\t\t: {gb_avg_f1:.2f}\n\n")

print(f"Decision TRee average AUC\t\t\t: {dt_avg_auc:.2f}")
print(f"Random Forest average AUC\t\t\t: {rf_avg_auc:.2f}")
print(f"Gradient Boosting average AUC\t\t\t: {gb_avg_auc:.2f}")
````

Like above, we'll directly perform hyper-parameeter tuning on max-depth, learning rate, number of estimators to use, and subsample fractions.

- max-depth : Like above, fixes the maximum depth of each tree. Deeper trees can cause overfitting.
- learning_rate : Step size of each epoch. Lower rate might take a long time to converge, but higher rate might not converge itself.
- n_estimators : Basically number of trees to build. This can balance overfitting and underfitting.
- subsample : Determines the fraction of samples to be used for each tree. Lower will reduce overfitting, but model can be conservative. 1 implies all samples are used for each epoch.

````python
param_grid = {
    #'booster': ['gbtree', 'gblinear', 'dart'],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'subsample': [0.8, 0.9, 1.0]
}
xgb_clf = xgb.XGBClassifier()

#We used GridSearch, but this is same as lopping across matrix and fitting the model .THis is optimised for multi-threading.
grid_search = GridSearchCV(estimator=xgb_clf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=0, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Best parameters : ", grid_search.best_params_)
print("Best accuracy : ", grid_search.best_score_)
````

We'll again see how  hyper-parameters are affecting the performance. Maybe exploit the trend by eye-balling to expand the grid.

````python
xgb_results = pd.DataFrame(grid_search.cv_results_)
xgb_results['n_estimators'] = xgb_results.params.apply(lambda x: x['n_estimators'])
xgb_results['learning_rate'] = xgb_results.params.apply(lambda x: x['learning_rate'])
xgb_results['max_depth'] = xgb_results.params.apply(lambda x: x['max_depth'])
xgb_results['subsample'] = xgb_results.params.apply(lambda x: x['subsample'])
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
sns.boxplot(ax=axes[0], data=xgb_results, x='n_estimators', y='mean_test_score')
axes[0].set_title('XGBoost: Accuracy vs N Estimators')
sns.boxplot(ax=axes[1], data=xgb_results, x='learning_rate', y='mean_test_score')
axes[1].set_title('XGBoost: Accuracy vs Learning Rate')
sns.boxplot(ax=axes[2], data=xgb_results, x='max_depth', y='mean_test_score')
axes[2].set_title('XGBoost: Accuracy vs Max Depth')
sns.boxplot(ax=axes[3], data=xgb_results, x='subsample', y='mean_test_score')
axes[3].set_title('XGBoost: Accuracy vs Subsample')
plt.tight_layout()
plt.show()
````

- Only learning rate seems to have some influence. Likely with 0.2, model is converging quickly.
- For number of estimators, adding more trees doesn't add any benefit.
- Accuracy doesnt change with depth of trees as well. But, a slight negative trend might indicate starting of overfitting with additional depth.

````python
xgb_model = xgb.XGBClassifier(max_depth=5, learning_rate=0.01, n_estimators=300, subsample=0.9)

scoring_metrics = ['accuracy', 'f1', 'roc_auc'] #metrics
xgb_scores = cross_validate(xgb_model, encoded_feats, binary_y, cv=k, scoring=scoring_metrics, return_train_score=True) #cross-validation

xgb_avg_test_accuracy = np.mean(xgb_scores['test_accuracy'])
xgb_avg_train_accuracy = np.mean(xgb_scores['train_accuracy'])
xgb_avg_f1 = np.mean(xgb_scores['test_f1'])
xgb_avg_auc = np.mean(xgb_scores['test_roc_auc'])

print(f"XGBoost average test accuracy\t: {xgb_avg_test_accuracy:.2f}")
print(f"XGBoost average train accuracy\t: {xgb_avg_train_accuracy:.2f}\n")
print(f"XGBoost average F1 score\t: {xgb_avg_f1:.2f}\n")
print(f"XGBoost average AUC\t\t: {xgb_avg_auc:.2f}")
````
