# Task 1 : CNN
For this task, we will implement a simple CNN model. We'll use it to classify images from CIFAR10 dataset (10 classes). We will implement the necessary functions to download the images and perform some transformations.

```python
import torch
import torch.nn as nn
import numpy as np
from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.tensorboard import SummaryWriter
#from torchsummary import summary
import matplotlib.pyplot as plt
import torch.nn.functional as F
import shutil
import torchvision as tv
import time
from torch.optim import Adam
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("Device ", device)
```

The data will downloaded through the pytorch APIs, and the transformations will be performed. Flags are passed as parameters, to modify it for Tasks 2 (resnet model) and 3 (data augmentation).

```python
label_names = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']


def get_cifar10_data_loaders(batch_size, model_name=None, validation_split=0.1, shuffle_dataset=True, random_seed=42, data_aug=False):
    """
    Downloads CIFAR10, performs necessary transformations, and creates data loaders for train, validate, and test splits,
    with specific adjustments if the model is ResNet-50 and optionally applies data augmentation.

    Args:
    - batch_size (int): Batch size for the data loaders.
    - model_name (str, optional): Model name to customize preprocessing. Supports "resnet50".
    - validation_split (float): Fraction of training data to be used for validation.
    - shuffle_dataset (bool): Whether to shuffle the dataset before splitting.
    - random_seed (int): Random seed for reproducibility.
    - data_aug (bool): Whether to apply data augmentation to the training dataset.

    Returns:
    - train_loader: DataLoader for the training set.
    - valid_loader: DataLoader for the validation set.
    - test_loader: DataLoader for the test set.
    """

    # basic transformations which is just converting it into tensor format
    base_transforms = [transforms.ToTensor()]

    if model_name == "resnet50": # normalization for resnet50 (comes from pytorch forums)
        base_transforms.append(transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))) # mean and std for resnet50
    else:  # general normalization
        base_transforms.append(transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)))

    # training and valdiation transformations with  data augmentation if needed
    train_transforms_list = []
    test_val_transforms_list = []

    if data_aug: # data augmentation
        train_transforms_list += [
            transforms.RandomHorizontalFlip(),  # randomly flip the image horizontally
            transforms.RandomRotation(10),  # randomly rotate the image by up to 10 degrees
        ]

    if model_name == "resnet50": # add resizing for ResNet-50. The standard model is defined for 224x224 pixels (ImageNet dataset)
        resize_transform = transforms.Resize(224) # resize the image to 224x224 pixels
        train_transforms_list.append(resize_transform) # add the resize transform to the list of transforms
        test_val_transforms_list.append(resize_transform)


    train_transforms_list.extend(base_transforms) # add the base transforms to the list of transforms
    test_val_transforms_list.extend(base_transforms)

    train_transforms = transforms.Compose(train_transforms_list) # combine all the transforms into a single transform
    test_val_transforms = transforms.Compose(test_val_transforms_list)

    # downloading/loading CIFAR10 dataset
    train_dataset = datasets.CIFAR10(root='data', train=True, download=True, transform=train_transforms)
    test_dataset = datasets.CIFAR10(root='data', train=False, download=True, transform=test_val_transforms)

    # indices for training and validation splits
    num_train = len(train_dataset)
    indices = list(range(num_train))
    split = int(np.floor(validation_split * num_train))

    if shuffle_dataset: # shuffle the indices
        np.random.seed(random_seed)
        np.random.shuffle(indices)

    train_indices, val_indices = indices[split:], indices[:split]

    train_sampler = SubsetRandomSampler(train_indices) #Sampler is mainly used to shuffle the indices, allows control for validation split
    valid_sampler = SubsetRandomSampler(val_indices)

    # creating data loaders for training, validation, and test sets
    train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=train_sampler) 
    valid_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=valid_sampler)
    test_loader = DataLoader(test_dataset, batch_size=batch_size)

    return train_loader, valid_loader, test_loader
```

```python
def imshow(img, labels): # helper function to show a grid of images
    plt.figure(figsize = (16,8))
    img = img / 2 + 0.5 # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.axis('off')
    # Grid dimensions
    grid_cols = 8
    grid_rows = 4
    img_width = npimg.shape[2] / grid_cols  # image width
    img_height = npimg.shape[1] / grid_rows  # image height

    for idx, label in enumerate(labels): # labels
        x = (idx % grid_cols) * img_width
        y = (idx // grid_cols) * img_height
        label_text = label_names[label]
        plt.text(x + 2, y + 2, label_text, color='white', backgroundcolor='black', fontsize=12, ha='left', va='top')
    plt.show()

def show_samples(train_load): # helper function to show a grid of images
    dataiter = iter(train_load)
    X, y = next(dataiter)  # get a batch from the iterator
    imshow(tv.utils.make_grid(X, nrow=8), y.tolist())
```

```python
train_load, valid_load, test_load = get_cifar10_data_loaders(batch_size)
show_samples(train_load)
```

Now, we'll define the model. It includes four convolutional layers, each followed by batch normalization and ReLU activations. Max pooling is performed after the first two convolutional layers to reduce dimensionality, and includes dropout for regularization. The model also has an adaptive average pooling layer to produce a fixed-size output, which is then fed into two FC layers. Output layer consists of the 10 classes.

```python
class CNN(nn.Module):
    def __init__(self, dropout_p=0.5):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 32, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(32)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(32, 64, 3, padding=1)
        self.bn3 = nn.BatchNorm2d(64)
        self.conv4 = nn.Conv2d(64, 64, 3, padding=1)
        self.bn4 = nn.BatchNorm2d(64)
        self.adaptive_pool = nn.AdaptiveAvgPool2d((5, 5))
        self.fc1 = nn.Linear(64 * 5 * 5, 512)
        self.bn5 = nn.BatchNorm1d(512)
        self.fc2 = nn.Linear(512, 10)
        self.dropout = nn.Dropout(dropout_p)

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.dropout(x)
        x = F.relu(self.bn3(self.conv3(x)))
        x = F.relu(self.bn4(self.conv4(x)))
        x = self.adaptive_pool(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = self.dropout(F.relu(self.bn5(self.fc1(x))))
        x = self.fc2(x)
        return x
```

```python
def train(cnn, optimizer, criterion, epoch, train_loader, device): # train the model for single epoch
    cnn.train()
    train_loss = 0
    correct = 0
    total = 0
    for batch_idx, (inputs, labels) in enumerate(train_loader):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = cnn(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

    avg_loss = train_loss / len(train_loader.dataset) # average training loss for the epoch
    accuracy = 100. * correct / total # calculate the accuracy for the epoch
    return avg_loss, accuracy
```

```python
def validate(cnn, criterion, valid_loader, device): # validation step
    cnn.eval()
    valid_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in valid_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = cnn(inputs)
            loss = criterion(outputs, labels)

            valid_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    avg_loss = valid_loss / len(valid_loader.dataset) #validation loss for the epoch
    accuracy = 100. * correct / total # calculate the accuracy for the epoch
    return avg_loss, accuracy
```

```python
cnn = CNN().to(device) 
criterion = nn.CrossEntropyLoss() # loss function
optimizer = Adam(cnn.parameters()) # optimizer for the model
```

```python
num_epochs = 30

# for storing metrics
train_accs = []
valid_accs = []

# Record the start time
start_time = time.time()

for epoch in range(num_epochs):
    # training
    train_loss, train_acc = train(cnn, optimizer, criterion, epoch, train_load, device)
    train_accs.append(train_acc)

    # validation
    valid_loss, valid_acc = validate(cnn, criterion, valid_load, device)
    valid_accs.append(valid_acc)

    tqdm.write(f'Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}, Valid Loss: {valid_loss:.4f}, Valid Acc: {valid_acc:.2f}')

total_time = time.time() - start_time
print(f'Training completed in: {total_time//60:.0f}m {total_time%60:.0f}s')
```

```python
plt.figure(figsize=(10, 6))
epochs = range(1, len(train_accs) + 1)
plt.plot(epochs, train_accs, label='Training Accuracy', marker='o')
plt.plot(epochs, valid_accs, label='Validation Accuracy', marker='o')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.xticks(epochs)
plt.show()
```

The model demonstrate a good performance after 30 epochs of training. The final validation accuracy comes to 82.8%, which is decent for a vanilla model. Both the training and validation accuracy has a good improvment inititally, and then plateus around 12th epoch.
